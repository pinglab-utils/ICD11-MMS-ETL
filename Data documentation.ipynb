# Data documentation

## ICD11 MMS Release date 2019-04-06

Our ICD11 MMS dataset consists of 33,577 JSON (JavaScript Object Notation) objects with the following fourteen properties (note: all of these properties are strings):



#### id

    The id used in actually retrieving this object from the api. Note: we sometimes use the following syntax,    id_other or id_unspecified but if you were to retrieve the same object from the ICD11 MMS API it would be
    id/other id/unspecified.

#### title

    The title field is in every JSON Object in this dataset. Essentially the name of the object.

#### code

    MMS(Mortality and Morbidity statistics) code, if not present in the original object this field is simply the empty string "". 

#### code_range

    String denoting the other MMS codes that this code can also describe, if not present in the original object it is simply the empty string "". The codes in this field are also child nodes of this code. Has the following syntax: code-code, to display the range.

#### class_kind

    Optional field, if not present in the original object it is simply the empty string "". 4 possible values:
    category, window, block, and chapter.

#### child_nodes

    Optional field, if not present in the original object it is simply the empty list/array []. A list of
    strings filled with the ids of all the child codes. 

#### definition

    Optional field, if not present in the original object it is simply the empty string "".

#### parent_nodes

    List of ids, if not present in the original object it is simply the empty list/array []. 

#### inclusions

    List of ids, if not present in the original object it is simply the empty list/array []. List is filled with all of the things you may want to also code for with this code.

#### exclusions

    Optional field, if not present in the original object it is simply the empty list/array []. A list of
    strings filled with all of the things to not code together with this code.

#### block_id

    String if not present in the original object it is simply the empty string "".

#### index_terms

    List of strings, if not present in the original object it is simply the empty list/array []. 
    
#### p_scale
    List of jsons with the following properties for each json in the list:
        allow_multiple_values 
                a string denoting whether this code allows for multiple values
        id 
            id of this object
        required_post_coord
            whether or not this
        scale_entity
            the id's of all the post coordination codes
    if not present in the original object this property is just an empty list.
